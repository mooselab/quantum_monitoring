```python
from qiskit.providers.fake_provider import FakeWashington
from qiskit import QuantumCircuit, Aer, execute, transpile
from qiskit.providers.aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.providers.aer.noise import NoiseModel
from qiskit.transpiler import passes
from qiskit.transpiler.passmanager import PassManager
import pathlib
from pathlib import Path
import numpy as np
import pandas as pd
from math import pi
import copy

import pickle
from qiskit.quantum_info import partial_trace, Statevector
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
```

```python
def find_probablity(file, qc, simulator):
    res = []
    partial_circuit = QuantumCircuit()
    for reg in qc.qregs:
        new_reg = QuantumRegister(reg.size, reg.name)
        partial_circuit.add_register(new_reg)
    # print(qc.num_qubits)
    qc = transpile(qc, simulator)
    for i, gate in enumerate(qc.data):
        # 创建一个新的电路，将所有先前的门加入到新的电路中
        # partial_circuit = QuantumCircuit(qc.num_qubits)
        partial_circuit.data = qc.data[:i+1]  # 只包含到第 i 步的操作
    
        # 模拟部分电路并获取态向量
        job = simulator.run(partial_circuit)
        result = job.result()
        statevector = result.get_statevector()
        
        # 将态向量转换为量子态对象
        quantum_state = Statevector(statevector)
    
        # 提取与该门操作相关的量子比特的状态
        qubits_involved = [partial_circuit.find_bit(qubit) for qubit in gate[1]]  # 获取与该门相关的量子比特
        for qubit in qubits_involved:
            # 使用 partial_trace 仅提取该量子比特的状态
            qubit_state = partial_trace(quantum_state, [q for q in range(qc.num_qubits) if q != qubit.index])
            prob_zero = qubit_state.data[0, 0].real
            prob_one = qubit_state.data[1, 1].real
            res.append([i, gate[0].name, qubit, prob_zero, prob_one])
    return res
```

```python
def main_function():
    f = open('../additional quantum circuits/statevector_list.pkl', 'ab+')

    simulator = Aer.get_backend('statevector_simulator')
    
    folder_path = Path("../additional quantum circuits/")  # 文件夹路径

    # result = {}

    # 递归遍历文件夹中的所有文件（包括子文件夹）
    for file in folder_path.rglob('*'):
        if file.is_file() and file.suffix == '.qasm':  # 检查文件是否以 .qasm 结尾
            print(file)
            try:
                qc = QuantumCircuit.from_qasm_file(file)
            except Exception as e:
                print(e)
                continue

            pm = PassManager(passes.RemoveFinalMeasurements())
            qc = pm.run(qc)

            res = find_probablity(file, qc, simulator)
            # print(res)
            # result[file] = res
            # print(result)
            # return
    
            pickle.dump([file, res], f)
            # break
    f.close()
            
    #         if find_def_state(qc, simulator):
    #             num += 1
            
    #         print(num)
    # return num
                
```

```python
def find_operated_gates(transpiled_qc, gate_list, start, index, qubit_info):
    insert_point = dict()
    for i in range(start, index):
        if qubit_info in [transpiled_qc.find_bit(qubit) for qubit in transpiled_qc.data[i][1]]:
            # gate_list.append(transpiled_qc.data[i])
            if transpiled_qc.data[i][0].name == 'swap':
                for second_qubit_info in [transpiled_qc.find_bit(qubit) for qubit in transpiled_qc.data[i][1]]:
                    if second_qubit_info != qubit_info:
                        # print(second_qubit_info)
                        gate_list = find_operated_gates(transpiled_qc, [], start, i, second_qubit_info)
                        # print(gate_list)
                        # print('\n')
                if not gate_list:
                    return []
                gate_list.append([i, transpiled_qc.data[i]])
                # print(gate_list)
                        
            else:
                # print(i)
                # print(gate_list)
                if qubit_info == transpiled_qc.find_bit(transpiled_qc.data[i][1][0]):
                    if transpiled_qc.data[i][0].num_qubits == 1:
                        gate_list.append([i, transpiled_qc.data[i]])
                    # gate_list.append([i, transpiled_qc.data[i]])
                elif transpiled_qc.data[i][0].num_qubits != 1 and qubit_info == transpiled_qc.find_bit(transpiled_qc.data[i][1][1]):
                    control_qubit_info = transpiled_qc.find_bit(transpiled_qc.data[i][1][0])
                    control_qubit_gate_list = []
                    for j in range(i):
                        if control_qubit_info in [transpiled_qc.find_bit(qubit) for qubit in transpiled_qc.data[j][1]]:
                            if transpiled_qc.data[j][0].num_qubits != 1:
                                # print(transpiled_qc.data[j])
                                if control_qubit_info == transpiled_qc.find_bit(transpiled_qc.data[j][1][0]):
                                    continue
                                else:
                                    # print(i, transpiled_qc.data[i])
                                    # return []
                                    # control_start = insert_point[control_qubit_info.registers[0]] \
                                    #     if control_qubit_info.registers[0] in insert_point else start
                                    # insert_point[control_qubit_info.registers[0]] = j
                                    control_qubit_gate_list = []
                                    control_control_qubit_gate_list = find_operated_gates(transpiled_qc, [], 0, i, control_qubit_info)
                                    if not control_control_qubit_gate_list:
                                        return []
                                    control_qubit_gate_list.extend(control_control_qubit_gate_list)
                                    break
                                    # print(find_operated_gates(transpiled_qc, [], i, control_qubit_info))
                            else:
                                # print(control_qubit_info)
                                control_qubit_gate_list.append([j, transpiled_qc.data[j]])
                                # print(transpiled_qc.data[j])
                                # print(control_qubit_gate_list)
                    gate_list.append([[i, transpiled_qc.data[i]], control_qubit_gate_list])
                else:
                    return []
    # print(gate_list)
    return gate_list
```

```python
def generate_new_operation_list(gate_list, node_index, gate_list_container):
    if node_index not in gate_list_container:
        gate_list_container[node_index] = []  # 为每个节点创建独立的映射
    for gate in gate_list:
        if gate[:4] in gate_list_container[node_index]:
            if len(gate) == 4:
                continue
        else:
            gate_list_container[node_index].append(gate[:4])
        
        if len(gate) == 5 and gate[4]:
            generate_new_operation_list(gate[4], node_index, gate_list_container)
        
        if len(gate) > 5:
            print('\n')
            print('\n')
            print('\n')
            print('\n')
            print('\n')
            print('red velvet')
            print('\n')
            print('\n')
            print('\n')
            print('\n')
            print('\n')
        
    return gate_list_container
```

```python
from qiskit.providers.fake_provider import FakeWashington
from qiskit import Aer, transpile, execute

qc = QuantumCircuit.from_qasm_file('../quantum circuits/qpeexact_indep_qiskit_5.qasm')
backend = Aer.get_backend('qasm_simulator')  # 使用模拟器
result = execute(qc, backend, shots=8192).result()
counts = result.get_counts()
print(counts)

# 使用 FakeWashington 模拟器
fake_washington = FakeWashington()

# 将量子电路转译为适应 FakeWashington
transpiled_qc = transpile(qc, backend=fake_washington)

# 使用 Aer 运行器执行电路
result = execute(transpiled_qc, backend=fake_washington, shots=8192).result()
print(result.get_counts())
```

```python
from qiskit import QuantumCircuit, transpile, Aer

# 创建量子电路
qc = QuantumCircuit(3,3)
qc.h(0)
qc.cx(0, 1)
# qc.measure(1, 0)  # 中途测量第二个qubit
qc.h(2)
qc.measure_all()

# 指定后端和编译电路
backend = Aer.get_backend('qasm_simulator')
tqc = transpile(qc, backend, basis_gates=['id', 'rz', 'sx', 'x', 'cx', 'reset'], optimization_level=1)


from qiskit.providers.fake_provider import FakeWashington, FakeTokyo
fake_washington = FakeWashington()
tqc2 = transpile(tqc, backend=fake_washington)

# 打印原始和编译后的电路
print("原始电路:")
print(qc.draw())

```
