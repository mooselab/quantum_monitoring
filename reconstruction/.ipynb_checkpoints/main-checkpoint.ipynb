# Get circuits with unexpected outputs and shifited distributions

```python
uneven_result_list = []
for er in execution_result:
    # actual_result_list = [k.split(' ')[-1] for k in list(list(er[0].items())[0][1][1].keys())]
    key = list(er[0].items())[0][0]
    if key not in operation_list_new:
        print(key)
        continue
    possible_result_list, possible_reconstructed_list = circuits_outputs[key]
    # print(key)
    # print(possible_result_list)
    # print(possible_reconstructed_list)
    # print('\n')
    if not set(possible_reconstructed_list).issubset(set(possible_result_list)) or list(er[0].items())[0][1][2] >= 0.1:
    # if list(er[0].items())[0][1][2] > 0.1:
        print(key)
        uneven_result_list.append(key)
```

# Get circuits with node whoes probabilities are changed

```python
with open('../quantum_circuits/circuits_monitoring_node_probability.pkl', 'rb') as file:
    circuits_monitoring_node_probability = pickle.load(file)

unexpected_result_list = []

count = 0
for key, value in circuits_monitoring_node_probability.items():
    print(key)
    flag = True
    for i in value:
        # print(i[-1])
        if (i[-1] and i[-1].split('_')[0] == 'test') and i[2][1][0][0].name != 'eq':
            index = int(i[-1].split('_')[1])
            # print(index)
            for r in operation_list_new[key]:
                # print(r)
                if r[0] == index:
                    if i[2].index != r[2].index:
                        continue
                    if abs(i[3] - r[3]) >= 1e-2 or abs(i[4] - r[4])>= 1e-2:
                        print(i[2], r[2], i[3], r[3], i[4], r[4])
                        flag = False
                        continue
    print('***********************', '\n')
    if not flag:
        count += 1
        unexpected_result_list.append(key)


print(count)
```

# Summary

```python
min_detect_count = 3
max_detect_count = 0
total_detect_count = 0
right_detect_count = 0
two_detect_count = 0
one_detect_count = 0
key_set = set()
for key, value in detection_result.items():
    mutation_id_list = []
    detect_count = 0
    for v in value:
        if v[0] in mutation_id_list:
            continue
        total_detect_count += 1
        if v[0] in v[1] or v[0] in list(v[1].values())[0]:
            detect_count += 1
            right_detect_count += 1
            key_set.add(key)
        mutation_id_list.append(v[0])
    if detect_count == 1:
        one_detect_count += 1
    if detect_count == 2:
        two_detect_count += 1
        # if v[0] not in v[1] and v[0] not in list(v[1].values())[0]:
        #     print(v)
    if detect_count > max_detect_count:
        max_detect_count = detect_count
    if detect_count < min_detect_count:
        min_detect_count = detect_count


print('min_detect_count = ', min_detect_count)
print('max_detect_count = ', max_detect_count)
print('right_detect_count = ', right_detect_count)
print('right_detect_circuit = ', len(key_set))
print('total_detect_count = ', total_detect_count)
print('one_detect_count = ', one_detect_count)
print('two_detect_count = ', two_detect_count)
print('total_detect_circuit = ', len(detection_result))
```
