```python
min_detect_count = 3
max_detect_count = 0
total_detect_count = 0
right_detect_count = 0
three_detect_count = 0
two_detect_count = 0
one_detect_count = 0
key_set = set()
TF = 0
TP = 0
mutation_set = set()
good_set = set()
mutation_detect_set = set()
bad_detect_set = set()
for key, value in detection_result.items():
    mutation_id_list = []
    detect_count = 0
    for v in value:
        if v[0] in mutation_id_list:
                continue
        if v[1] == 'P':
            good_set.add((key, v[0]))
            if abs(v[4][0] - v[4][1]) >= 1e-2 or abs(v[5][0] - v[5][1]) >= 1e-2:
                bad_detect_set.add((key, v[0]))
                mutation_detect_set.add((key, v[0]))
                if v[0] in v[2] or v[0] in list(v[2].values())[0]:
                    detect_count += 1
                    right_detect_count += 1
                    key_set.add(key)
                    mutation_id_list.append(v[0])
        else:
            mutation_set.add((key, v[0]))
            # total_detect_count += 1
            if abs(v[4][0] - v[4][1]) >= 1e-2 or abs(v[5][0] - v[5][1]) >= 1e-2:
                mutation_detect_set.add((key, v[0]))
                if v[0] in v[2] or v[0] in list(v[2].values())[0]:
                    detect_count += 1
                    right_detect_count += 1
                    key_set.add(key)
                    mutation_id_list.append(v[0])
    if detect_count == 1:
        one_detect_count += 1
    if detect_count == 2:
        two_detect_count += 1
    if detect_count == 3:
        three_detect_count += 1
        # if v[0] not in v[1] and v[0] not in list(v[1].values())[0]:
        #     print(v)
    if detect_count > max_detect_count:
        max_detect_count = detect_count
    if detect_count < min_detect_count:
        min_detect_count = detect_count


print('min_detect_count = ', min_detect_count)
print('max_detect_count = ', max_detect_count)
print('right_detect_count = ', right_detect_count)
print('right_detect_circuit = ', len(key_set))
print('total_detect_count = ', total_detect_count)
print('one_detect_count = ', one_detect_count)
print('two_detect_count = ', two_detect_count)
print('three_detect_count = ', three_detect_count)
print('total_detect_circuit = ', len(detection_result))
print('len(mutation_detect_set) =' , len(mutation_detect_set))
print('len(bad_detect_set) =' , len(bad_detect_set))
print('len(mutation_set) =' , len(mutation_set) + len(bad_detect_set))
print('len(good_set) =' , len(good_set) - len(bad_detect_set))
```
